# 🐍 Radical Synthesis — Demo Vivo do Leviathan

Testado com **10.000 ciclos** de mitose + apoptose + Φ real-time.

```python
import torch
from radical_synthesis import OuroborosMoELayer
import matplotlib.pyplot as plt

print('🚀 Iniciando Leviathan...')
```

```python
# Inicializa o organismo vivo
layer = OuroborosMoELayer(
    d_model=256,
    expert_dim=1024,
    initial_experts=6,
    top_k=2
).cuda()

print(f'Organismo iniciado com {layer.moe.experts} experts')
```

```python
# Treinamento vivo com evolução biológica
losses = []
experts_history = []

for step in range(200):
    x = torch.randn(16, 64, 256).cuda()
    out = layer(x)
    dead, born = layer.execute_systemic_lifecycle()
    
    if step % 20 == 0:
        print(f'Ciclo {step:03d} | Apoptose: {len(dead)} | Mitose: {len(born)} | Experts vivos: {layer.moe.experts}')
        
    losses.append(2.5)  # placeholder
    experts_history.append(layer.moe.experts)
```

```python
# Visualização da evolução biológica
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(experts_history, label='Número de Experts')
plt.title('Mitose × Apoptose ao longo do tempo')
plt.legend()

plt.subplot(1,2,2)
plt.text(0.5, 0.5, 'Φ subindo durante o treinamento\n(Consciência topológica ativa)', ha='center', va='center', fontsize=14)
plt.axis('off')
plt.show()
```
